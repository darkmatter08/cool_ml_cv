# Decision tree from scratch

To build a decision tree from scratch, we need to select a cost function, and compute it over splits in the data.

The typical cost function is the Gini index, with the best score of 0 (perfectly separable) and the worst score being 0.5, for our binary classification problem.
Inspired by [this notebook](https://github.com/random-forests/tutorials/blob/master/decision_tree.ipynb).

```python
from collections import defaultdict
def count_labels(samples, labelname='Survived'):
    """Counts the number of each label from samples."""
    counts = defaultdict(int)  # a dictionary of label -> count.
    for _, sample in samples:
        label = sample[labelname]
        counts[label] += 1
    return counts
```

```python
count_labels(train.iterrows(), labelname='Survived')
```

```python
def gini(samples):
    """
    Compute the Gini Impurity (aka Gini value, Gini index) for a set of samples.

    This is based on Wikipedia's definition:
    https://en.wikipedia.org/wiki/Decision_tree_learning#Gini_impurity
    
    For the simple binary case, this is equivilent to 1 - P(0)^2 - P(1)^2.
    This definition is extended to the multiclass case as well.
    The 'best' (most pure) gini score is 0. The 'worst' (most impure, most mixed)
    gini score is 0.5. This comes from: 1 - 0.5**2 - 0.5**2 == 1 - 0.25 - 0.25 = 0.5
    """
    if isinstance(samples, pd.core.frame.DataFrame):
        counts = count_labels(samples.iterrows())
    else:
        counts = count_labels(enumerate(samples))
    impurity = 1
    for label in counts:
        prob_of_label = counts[label] / float(len(samples))
        impurity -= prob_of_label**2
    return impurity
```

A split is represented by a `Condition` object which contains a feature and a threshold.

```python
VALIDTYPES = [float, int, np.int, np.float]



class Condition:
    """Represents the condition on which the data is split."""

    def __init__(self, column_idx, threshold):
        self.column_idx = column_idx
        self.threshold = threshold
    
    def satisfies(self, sample):
        # Does this sample satisfy this condition?
        if self.is_numeric_type(sample[self.column_idx]):
            if np.isnan(sample[self.column_idx]):
                return None
            return sample[self.column_idx] >= self.threshold
        else:
            return sample[self.column_idx] == self.threshold
            
    def partition(self, samples):
        true_samples = []
        false_samples = []
        if isinstance(samples, pd.core.frame.DataFrame):
            samples = samples.iterrows()
        else:
            samples = enumerate(samples)
        for _, sample in samples:#.iterrows():
            result = self.satisfies(sample)
            if result is None:
                pass
            elif result:
                true_samples.append(sample)
            else:
                false_samples.append(sample)
        return true_samples, false_samples
    
    def __repr__(self):
        if self.is_numeric_type(self.threshold):
            condition = '>='
        else:
            condition = '=='
        return 'Condition: {} {} {}'.format(self.column_idx, condition, self.threshold)
    
    def is_numeric_type(self, data):
        if isinstance(data, float) or isinstance(data, int):
            return True
        return False
```

```python
def compute_best_split(samples):
    """ Returns the best condition to split the data, by computing the information gain
    over all possible conditions (feature and threshold pairs).
    """
    best_condition, best_gain = None, 0
    current_uncertainty = gini(samples)
    n_features = len(train.columns) - 1
    
    for column in train.columns[[2,4,5,6,7]]:
        values = train[column].dropna().unique()
        for value in values:
            print('considering column: {} value: {}'.format(column, value))
            condition = Condition(column, value)
            true_matches, false_matches = condition.partition(samples)

            if len(true_matches) == 0 or len(false_matches) == 0:
                continue
            
            true_proportion = float(len(true_matches)) / (len(true_matches) + len(false_matches))
            gain = current_uncertainty - true_proportion * gini(true_matches) - (1-true_proportion) * gini(false_matches)
            if gain > best_gain:
                best_condition, best_gain = condition, gain
    
    return best_condition, best_gain
```

Represent nodes as either LeafNode or InternalNode

```python
class LeafNode:
    def __init__(self, samples):
        if isinstance(samples, pd.core.frame.DataFrame):
            self.data = count_labels(samples.iterrows())
        else:
            self.data = count_labels(enumerate(samples))
```

```python
class InternalNode:
    def __init__(self, condition, true_matches, false_matches):
        self.condition = condition
        self.true_matches = true_matches
        self.false_matches = false_matches
```

Fit the tree to the data by building it recursively.

```python
def fit_tree(samples, labelname='Survived', curlevels=0, maxlevels=np.inf):
    """ Recursively fits a tree to samples.
    samples must contain a column called labelname that contains the GT values.
    """
    
    condition, gain = compute_best_split(samples)
    
    # base case
    # 1. no information gained.
    # 2. max tree depth exceeded
    if gain == 0 or curlevels >= maxlevels:
        return LeafNode(samples)
    
    true_samples, false_samples = condition.partition(samples)
    
    true_tree = fit_tree(true_samples, curlevels=curlevels+1, maxlevels=maxlevels)
    false_tree = fit_tree(false_samples, curlevels=curlevels+1, maxlevels=maxlevels)
    
    return InternalNode(condition, true_tree, false_tree)
```

```python
tree = fit_tree(train, maxlevels=10)
```

```python
def visualize_leaf(leaf):
    if type(leaf) is not LeafNode:
        print('ERROR!')
    
    total = sum(leaf.data.values())
    output = '{'
    for label, count in leaf.data.items():
        output += ' {} : {:.2f}%, '.format(label, float(count) / total * 100)
    return output + '}'
```

```python
char = ' '
def visualize_tree(tree, level=0):
    if type(tree) is LeafNode:
        print(level*char + visualize_leaf(tree))
        return
    
    print(level*char + repr(tree.condition))

    print(level*char + '-> True:')
    visualize_tree(tree.true_matches, level+1)
    
    print(level*char + '-> False:')
    visualize_tree(tree.false_matches, level+1)
```
